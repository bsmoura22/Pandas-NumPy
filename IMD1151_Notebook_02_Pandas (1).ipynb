```python
import pandas as pd
```

```python
ufrn_consumo = pd.read_csv("consumo-energetico-por-campus.csv", sep = ';')
```

```python
type(ufrn_consumo)
```

```python
ufrn_consumo.head(12)
```

```python
ufrn_consumo.dtypes
```

```python
ufrn_consumo['ano']
```

```python
ufrn_consumo["ano"] = ufrn_consumo["ano"].astype(int)
ufrn_consumo['numero_fatura'] = ufrn_consumo['numero_fatura'].astype(int)
ufrn_consumo["mes_referencia_fatura"] = ufrn_consumo["mes_referencia_fatura"].astype(int)
ufrn_consumo.dtypes
```

```python
ufrn_consumo['valor_fatura'] = ufrn_consumo['valor_fatura'].str.replace(',', '.')
ufrn_consumo["valor_fatura"] = ufrn_consumo["valor_fatura"].astype(float)

ufrn_consumo['quantidade_conceito_faturado'] = ufrn_consumo['quantidade_conceito_faturado'].str.replace(',', '.')
ufrn_consumo["quantidade_conceito_faturado"] = ufrn_consumo['quantidade_conceito_faturado'].astype(float)

ufrn_consumo['valor_conceito_faturado'] = ufrn_consumo['valor_conceito_faturado'].str.replace(',', '.')
ufrn_consumo["valor_conceito_faturado"] = ufrn_consumo['valor_conceito_faturado'].astype(float)


ufrn_consumo.dtypes
```

```python
ufrn_consumo['cep'] = ufrn_consumo['cep'].str.replace(' ', '')
ufrn_consumo['cep'] = ufrn_consumo['cep'].str.replace('A', '')
ufrn_consumo["cep"] = ufrn_consumo["cep"].astype(int)
```
